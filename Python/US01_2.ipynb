The goal is to feel an xlsx with the GDP of a list of countries. The list of contries are on a datasheet. It's possible to get the gdp on internet thanks a API  (application programming interface). 


**1. List of the countries**

We need to get the list of countries from the datasheet. We import the data of the datasheet in a dataframe

```python
#import a library to open xlsx
import pandas as pd 
```

```python
#read_excel read the xlsx. It contains many parameters. 
#see more: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_excel.html
df = pd.read_excel('Data-sheet-STUDENTS.xlsx',
             header=2,
             usecols=[1],
             names=['countries'] 
             )
```

```python
df['countries']
```

**2. Get the ISO code**

We need know to get the GDP thanks to a API: the API world bank accept the ISO code and not the properly name of the country. So we need to change the name of country into the ISO code. For example: France = fr

To do so we can use this API : https://restcountries.eu/rest/v2/name/france?format=json and a dictionnary which contains all the countries from our datasheet

```python
countries={}
for country in df['countries'] :
    countries[country] = {}
countries
```

```python
import requests
from time import sleep
#from tqdm import tnrange, tqdm_notebook

#from i in tnrange(len(countries.keys())):
for country in countries.keys():
    try: 
        #counrty = list(countries.keys())[i]
        if countries[country] == {}: 
            url = ' https://restcountries.eu/rest/v2/name/'+country
            r = requests.get(url)
            data = r.json();
            countries[country]['alpha2Code'] = data[0]['alpha2Code']
        
    except KeyError: 
        print(data, country)
        sleep(5)
```

We can see that there is a problem for 6 country : {'status': 404, 'message': 'Not Found'}. We need to dig more futher to find why it doesn't work, find a solution and the ISO code.

```python
for country, code in countries.items():
    countryCorrect = ''
    if code == {}:
        if 'not bri' in country.lower(): #france, germany, united kingdom, united states of america
            countryCorrect  = country.split('(')[0].strip()
        elif country == 'Thailiand': #thailand
            countryCorrect = 'Thailand'
        else: #the united arab emirates
            countryCorrect = country[4:]
        try: 
        #counrty = list(countries.keys())[i]
            if countries[country] == {}: 
                url = ' https://restcountries.eu/rest/v2/name/'+countryCorrect
                r = requests.get(url)
                data = r.json();
                countries[country]['alpha2Code'] = data[0]['alpha2Code']
        
        except KeyError: 
            print(data, country)
            sleep(5)    

            
```

So now, know, we have all our countries name with their ISO code in our dictionnary of countries:

```python
countries
```

We can see that the ISO code for India isn't right (IO), so we change it manually. 

```python
countries['India']['alpha2Code'] = 'IN'
```

**3. GDP for each country**

We need to use another API to get the GDP from world bank thanks to the ISO code. 

```python
for k,v in countries.items():
    try: 
        url = 'http://api.worldbank.org/v2/countries/'+v['alpha2Code']+'/indicators/NY.GDP.MKTP.CD?format=json'
        r = requests.get(url)
        data = r.json()
        countries[k]['gdp'] = float(data[1][0]['value'])
        sleep(2)
    except TypeError: 
        print('typeerror: ', k)        
    except IndexError: 
        print('indexerror: ', k)    
        
countries        
```

We can see that Yemen and Syria don't have a GPD (because of the war). So we can add a GDP at "None"

```python
countries['Yemen']['gdp'] = 'None'
countries['Syria']['gdp'] = 'None'
```

**4. Add the gdp value into the datasheet**

First we add the GDP in the dataframe. Secondly, we have to add into the datasheet

```python
#add into the dataframe
s = pd.Series([v['gdp'] for k,v in countries.items()])
df['gdp'] = s
df
```

```python
#add a graph from the dataframe
%matplotlib inline
import matplotlib.pyplot as plt

df.plot()
```

```python
#add into the datasheet
writer = pd.ExcelWriter('output.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
